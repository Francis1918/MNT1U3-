```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# Definir el rango de x
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Función original
def original_function(x):
    return np.cos(x)

# Función para calcular el polinomio de Taylor de orden n para cos(x) en x0=0
def taylor_cos(x, n):
    result = 0
    for i in range(n+1):
        # Solo los términos pares aparecen en la serie de cos(x)
        if i % 2 == 0:
            # El signo alterna: 1, -1, 1, -1, ...
            sign = (-1)**(i//2)
            result += sign * (x**i) / factorial(i)
    return result

# Crear la gráfica
plt.figure(figsize=(12, 8))

# Graficar la función original
plt.plot(x, original_function(x), 'k-', label='cos(x)', linewidth=2)

# Graficar los polinomios de Taylor de diferentes órdenes
orders = [2, 4, 6, 10]
colors = ['r', 'g', 'b', 'm']

for i, order in enumerate(orders):
    taylor_approx = taylor_cos(x, order)
    plt.plot(x, taylor_approx, f'{colors[i]}--',
             label=f'Taylor orden {order} en $x_0=0$')

# Configurar la gráfica
plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='gray', linestyle='-', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.title('Función cos(x) y sus aproximaciones de Taylor en $x_0=0$')
plt.xlabel('x')
plt.ylabel('y')

# Limitar el rango del eje y para mejor visualización
plt.ylim(-2, 2)

# Añadir marcas para múltiplos de π en el eje x
ticks = np.arange(-2*np.pi, 2*np.pi + 0.1, np.pi/2)
labels = [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$',
          r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']
plt.xticks(ticks, labels)

plt.show()

# Mostrar las fórmulas de los primeros polinomios de Taylor
print("Polinomios de Taylor para cos(x) en x₀ = 0:")
print("Orden 2: cos(x) ≈ 1 - x²/2")
print("Orden 4: cos(x) ≈ 1 - x²/2 + x⁴/24")
print("Orden 6: cos(x) ≈ 1 - x²/2 + x⁴/24 - x⁶/720")
print("Orden 10: cos(x) ≈ 1 - x²/2 + x⁴/24 - x⁶/720 + x⁸/40320 - x¹⁰/3628800")
```
